```python
import pickle
from matplotlib import pyplot as plt
import numpy as np
import pickle
import glob

data_folder='/arc/projects/MachineLearning/Hossen/data/JJ/'
lst= glob.glob(data_folder+'/*_med*')
tp=[]

for k1 in range(len(lst)):

    temp= pickle.load(open(lst[k1], 'rb'))
    tp.append(temp)

for k1 in range(len(lst)):
    if k1 ==0:tpt=np.array(tp[k1])
    else:
        tpt= np.concatenate((tpt,tp[k1]),axis=0)
print (tpt.shape)

tpt_1d= np.reshape(tpt,(-1,441))
pr= np.percentile(tpt_1d,[1,99],axis=1)

tp_M=[]
tp_S=[]
for k1 in range(len(tpt)):
    if k1%10000==0:print(k1,end=', ')
    temp=tpt_1d[k1]
    temp[temp<pr[0,k1]]= pr[0,k1]
    temp[temp>pr[1,k1]]= pr[1,k1]
    mu = np.mean(temp)
    sig= np.std(temp)
    min_= np.min(temp)
    max_ = np.max(temp)
    if min_==0:min_ = min_+.0001
    temp_m= (temp-min_)/(max_-min_)
    temp_s= (temp-mu)/(sig)
    tp_M.append(temp_m)
    tp_S.append(temp_s)
        

tp_M= np.array(tp_M)
tp_S= np.array(tp_S)

np.save (data_folder+'images_med_Norm_M', tp_M)
np.save (data_folder+'images_med_Norm_S', tp_S)
```
